# DATATYPES 

--> Datatype will specify the size and type of value stored in a variable.

--> Based on size of the value datatype got classified into 2 categories:
   
    1) Single value / Individual Datatype
    2) Multi value  / Collection Datatype

--> When we store a single value in a single variable then it is called as
    Single value / Individual Datatype.

--> When we store a multiple values in a single variable then it is called 
    as Multi value / Collection Datatype.


## 01) integer (int) :

--> It is a real number without decimal point

--> Integer can have both positive and negative values.

--> Each and every datatype will be having Two types of valus:
    1) default value
    2) Non default value
    
--> default value is an initial value which is internally equals to False.

--> Other than default value all the values are considered as Non default
    values , which is internally equals to True.
    
--> default value of integer is Zero (0)

--> Non default value of integer are 1,-50,4,68,777 etc

Example : a=10

## 05) String (str) :-

--> It is a collection of characters enclosed between ' ' or " " or ''' '''

Syntax :

           var = 'val1val2val3.....valn'
                    (or)
           var = "val1val2val3.....valn"
                    (or)
           var = '''val1val2val3.....valn'''

--> In string we can store Alphabets,digits,special characters.

--> In string there is no seperation between the values.

--> default value for string is ''

--> If we create string with ' ' or " " or ''' ''' , internally it will be
    stored in the form of ' '.

--> " " string is used when ' ' is already present inside the string.

--> ''' ''' string is also called as doc string, which is used as comment,
    to write multiple lines.

--> If we create string with ' ' or  " " or  ''' ''' , we need to close it 
    with same type.

Example:   s = 'kerala'

Memory Allocation :-





--> After creating a collection , controller will create a layer of memory in 
    value space which will get divided into number of blocks that is equals
    to the length of the collection.

--> Then each and every value from given collection will get stored into each
    and every block sequentially.

--> Interpreter will generate one reference address for the layer of memory
    which will get stored into variable space with respect to variable name.
