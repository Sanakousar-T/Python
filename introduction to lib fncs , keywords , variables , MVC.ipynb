<h1>VARIABLE :</h1>

It is a name given to the memory location where the value is stored
                  (OR)
Variable is a container which is used to store the value.

<h3>Syntax:</h3>
          var_name  =  value

<h3>Memory Allocation for variable creation :</h3>

--> After creating a variable,internally memory allocation will 
    happen for that variable and the memory will get divided into 
    two parts : 
    1) variable space/stack space 
    2) value space/heap space

--> Value will get stored into value space and interpreter will generate 
    one reference address for that value , which will be stored inside
    variable space with respect to the particular variable name.


Ex: a=10
